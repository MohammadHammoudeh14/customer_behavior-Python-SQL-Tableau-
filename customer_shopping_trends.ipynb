```python
import pandas as pd
```

```python
path = '/content/drive/MyDrive/Data /customer_shopping_behavior.csv'
```

```python
df = pd.read_csv(path)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
# Filling each null vlue with the category median value
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
# Standardize column names: convert to lowercase and replace spaces with underscores for easier access
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
df.columns
```

```python
# create a column age_group
labels = ['Young Adult','Adult','Middle-Aged','Senior']
df['age_group'] = pd.qcut(df['age'],q=4, labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
# create columns purchase_frequency_days
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head()
```

```python
# inspect discount_applied and promo_code_used columns
df[['discount_applied','promo_code_used']].head(10)
```

```python
# check the similarity between the two columns
df['discount_applied'].equals(df['promo_code_used'])
```

```python
# they are the same 100%, so i will drop promo code column
df = df.drop('promo_code_used',axis=1)
```
